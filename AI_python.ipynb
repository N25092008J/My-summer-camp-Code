Linear Regression

Diabetes model



For importing the libraries from Scikit learn

```
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score 
```

For Loading the Datasets

```
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)
```

For checking the data

```
print(diabetes_X)
```

```
print(diabetes_y)
```

Extracting average bp (third column)

```
# Use only one feature
diabetes_X = diabetes_X[:, np.newaxis, 2]
```

```
print(diabetes_X)
```

For splitting data into training and testing

X is independent variable

```
# Split the data into training/testing sets
diabetes_X_train = diabetes_X[:-20]
diabetes_X_test = diabetes_X[-20:]
```

```
print(diabetes_X_test)
```

```
print(diabetes_X_train)
```

y is dependent variable

```
# Split the targets into training/testing sets
diabetes_y_train = diabetes_y[:-20]
diabetes_y_test = diabetes_y[-20:]
```

```
print(diabetes_y_test)
```

```
print(diabetes_y_train)
```

Now that we have training and test data we are going to create a model

```
# Create linear regression object
regr = linear_model.LinearRegression()
```

For Training the model

```
# Train the model using the training sets
regr.fit(diabetes_X_train, diabetes_y_train)
```

For predicting

```
# Make predictions using the testing set
diabetes_y_pred = regr.predict(diabetes_X_test)
```

For Data Analysis

Scatter plot (Test Data) and Line plot with predict data

```
plt.scatter(diabetes_X_test, diabetes_y_test,  color='black')
# line plot with predicted data
plt.plot(diabetes_X_test, diabetes_y_pred, color='blue', linewidth=3)
# for not having values in the x axis and y axis
plt.xticks(())
plt.yticks(())

plt.show()
```

```
# The coefficients
print('Coefficients: \n', regr.coef_)
```

```
# for MSE
mse=mean_squared_error(diabetes_y_test, diabetes_y_pred)
print(mse)
```

```
# for r2 score
r2score=r2_score(diabetes_y_test, diabetes_y_pred)
print(r2score)
```
